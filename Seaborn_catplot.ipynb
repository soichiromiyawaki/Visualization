# Plotting with categorical data

Categorical scatterplots:  
- stripplot() (../generated/seaborn.stripplot.html#seaborn.stripplot) (with kind="strip" ; the default)  
- swarmplot() (../generated/seaborn.swarmplot.html#seaborn.swarmplot) (with kind="swarm" )   

Categorical distribution plots:  
- boxplot() (../generated/seaborn.boxplot.html#seaborn.boxplot) (with kind="box" )
- violinplot()(../generated/seaborn.violinplot.html#seaborn.violinplot) (with kind="violin" ) 
- boxenplot()(../generated/seaborn.boxenplot.html#seaborn.boxenplot) (with kind="boxen" )  

Categorical estimate plots:  
- pointplot() (../generated/seaborn.pointplot.html#seaborn.pointplot) (with kind="point" )   
- barplot()   (../generated/seaborn.barplot.html#seaborn.barplot) (with kind="bar" )   
- countplot() (../generated/seaborn.countplot.html#seaborn.countplot) (with kind="count" )  

```python
import seaborn as sns
import matplotlib.pyplot as plt 
sns.set(style="ticks", color_codes=True)
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.catplot(x='day', y='total_bill', data=tips);
```

```python
#relplot()との違い
sns.relplot(x='day', y='total_bill', data=tips);
```

```python
sns.catplot(x='day', y='total_bill', data=tips, jitter=False);
```

```python
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips);
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips);
```

## 箱ひげグラフ

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box');
```

```python
sns.boxplot(x='day', y='total_bill', data=tips);
```

```python
g=sns.catplot(x='day', y='total_bill', data=tips, kind='box', hue='smoker')
g.fig.suptitle('USE suptitle for title');
```

sns.catplot(x='day', y='total_bill', data=tips, kind='boxen');

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin');
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='sex', split=True);
```

# Bar plots

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
sns.catplot(x='sex', y='survived', data=titanic, kind='bar');
```

```python
g = sns.catplot(x='sex', y='survived', data=titanic, kind='bar', hue='class')
g.fig.suptitle('Title');
```

## Point plots

```python
g = sns.catplot(x="sex", y="survived", hue="class", kind="point", data=titanic)
g.fig.suptitle('Title');
```

```python
fig, axes = plt.subplots(2,1, figsize=(10,7))

ax1 = axes[0]

sns.catplot(x='sex', y='survived', data=titanic, kind='bar', ax=ax1)

plt.show()
```
