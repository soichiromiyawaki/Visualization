```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from scipy import stats
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips);
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker');
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', style='smoker');
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, size='size');
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, size='size', sizes=(10,200), hue='size');
```

```python
time=np.arange(500)
time
```

```python
df = pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
df.head()
```

```python
g = sns.relplot(x='time', y='value', data=df, kind='line')
g.fig.autofmt_xdate();
```

```python
df = pd.DataFrame(np.random.randn(500, 2).cumsum(axis=0), columns=["x", "y"])
df.head()
```

```python
sns.relplot(x="x", y="y",  kind="line", data=df);
```

```python
sns.relplot(x="x", y="y", sort=False, kind="line", data=df);
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri, ci=None);
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri, ci='sd');
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri, estimator=None);
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri, hue='event');
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri, hue='region', style='event');
```
