## plt.subplots()
is a convinence method that creates a new figure and returns a NumPy array containing the created subplot objects

```python
fig, axes = plt.subplots(2,3, figsize=(12,6))
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,6))

axes[0, 0].hist(np.random.randn(500), bins=50, alpha=0.5)
axes[0, 1].bar(x, y, alpha=0.5)

plt.show()
```

```python
fig, axes = plt.subplots(3,1)

data.plot.bar(ax=axes[0])
data.plot.hist(ax=axes[1])
data.plot.barh(ax=axes[2])

plt.show()
```

# Simple Plot

```python
movies_per_month = [1, 2, 4, 6, 5, 3, 2, 4, 4, 6, 7, 9]
month = range(1,13)

plt.plot(month, movies_per_month)
plt.show()
```

```python
movies_per_month = [1, 2, 4, 6, 5, 3, 2, 4, 4, 6, 7, 9] 
jokes_per_month = [0, -1, 1, 3, 2, 2, 3, 4, 7, 9, 10, 11]

plt.plot(month, movies_per_month)
plt.plot(month, jokes_per_month)

plt.show()
```

### Legend

```python
plt.plot(month, movies_per_month)
plt.plot(month, jokes_per_month)

#Add a Legend
plt.legend(['Movie', 'Jokes'], loc=0)

plt.show()
```

### Title

```python
plt.plot(month, movies_per_month)
plt.plot(month, jokes_per_month)

#Add a Legend
plt.legend(['Movie', 'Jokes'], loc=0)

#Add a Title
plt.title('Simple Plot Title')

plt.show()
```

### Grid

```python
plt.plot(month, movies_per_month)
plt.plot(month, jokes_per_month)

#Add a grid
plt.grid(True)

#Add a Legend
plt.legend(['Movie', 'Jokes'], loc=0)

#Add a Title
plt.title('Simple Plot Title')

plt.show()
```
